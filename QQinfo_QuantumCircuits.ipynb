```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram
```

```python
# Create a quantum circuit with 1 qubit

circuit = QuantumCircuit(1)
circuit.h(0)    # Apply Hadamard gate on qubit 0
circuit.s(0)    # Apply S gate on qubit 0
circuit.h(0)    # Apply another Hadamard gate on qubit 0
circuit.t(0)    # Apply T gate on qubit 0

# Draw the circuit using Matplotlib
circuit.draw(output="mpl")
```

```python
# Create a quantum register with 1 qubit named 'X'

X = QuantumRegister(1, 'X')

# Create a quantum circuit using the quantum register
circuit = QuantumCircuit(X)

# Add gates to the quantum circuit
circuit.h(X)  # Apply Hadamard gate on qubit X
circuit.s(X)  # Apply S gate on qubit X
circuit.h(X)  # Apply another Hadamard gate on qubit X
circuit.t(X)  # Apply T gate on qubit X

# Draw the circuit using Matplotlib
circuit.draw(output="mpl")
```

```python
# Create quantum registers with 1 qubit each, named 'X' and 'Y'
X = QuantumRegister(1, 'X')
Y = QuantumRegister(1, 'Y')

# Create classical registers with 1 bit each, named 'A' and 'B'
A = ClassicalRegister(1, 'A')
B = ClassicalRegister(1, 'B')

# Create a quantum circuit using the quantum and classical registers
circuit = QuantumCircuit(Y, X, B, A)

# Add gates to the quantum circuit
circuit.h(Y)         # Apply Hadamard gate on qubit Y
circuit.cx(Y, X)     # Apply CNOT gate with control qubit Y and target qubit X
circuit.measure(Y, B)  # Measure qubit Y and store the result in classical bit B
circuit.measure(X, A)  # Measure qubit X and store the result in classical bit A


circuit.draw(output="mpl")
```

```python
# Initialize the Sampler and run the circuits to get the result
result = Sampler().run(circuit).result()


# Extract the quasi-probability distribution for the first circuit
# The quasi-probability distribution contains probabilities (which may include negative values) for each measurement outcome
# "Binary_probabilities()": Converts the quasi-probability distribution into a standard probability distribution
statistics = result.quasi_dists[0].binary_probabilities()


# Plot and display the histogram of the binary probability distribution
# This histogram visualizes the probabilities of different measurement outcomes
display(plot_histogram(statistics))
```
